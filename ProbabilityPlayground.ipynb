```python
import numpy as np
import matplotlib.pyplot as plt
```

# Coin and dice flip Simulation

```python
def sim_coins(size):
    return np.random.randint(0, 2, size=size)

def dice_roll(size, d_num):
    return np.random.randint(1, d_num+1, size=size)
```

```python
# Verify Law of Large Numbers, plotting ratio between heads and tails
# Should avg out to 1 over time

ratios = []
for i in np.linspace(5, 400, 400-5):
    flips = sim_coins(int(i))
    #print(flips)
    heads = 0
    tails = 0
    for j in flips:
        if j == 1:
            heads += 1
        else:
            tails += 1
    ratios.append(heads/tails)

plt.plot(ratios)
plt.plot(np.ones(400-5))
```

```python
# Verify CLT

trails = 10000
num_dice = 50

means = []
for i in range(trails):
    rolls = dice_roll(num_dice, 6)
    means.append(np.mean(rolls))

plt.figure(figsize=(10, 6))
plt.hist(means, bins=30, density=True, alpha=0.7)

mu = 3.5
sigma = np.sqrt(2.9167 / num_dice)
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
plt.plot(x, 1/(sigma * np.sqrt(2*np.pi)) * np.exp(-(x - mu)**2 / (2*sigma**2)), color='red', linewidth=2, label='Theoretical Normal Curve')

plt.title(f'Distribution of Means from {trails} Trials ({num_dice} Dice/Trial)', fontsize=14)
plt.xlabel('Sample Mean Value', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()
```

# Monte Carlo to estimate pi

```python
# make a square (-1,-1) to (1,1)

r = 1

# generating points
x = np.random.randint(-100,100, 10000) * (1/100)
y = np.random.randint(-100,100, 10000) * (1/100)

inside = (x**2 + y**2) <= r
theta = np.linspace(0, 2 * np.pi, 100)

x_c = r * np.cos(theta)
y_c = r * np.sin(theta)

plt.plot(x_c, y_c, color='red')
plt.scatter(x,y)

pi_est = 4 * np.mean(inside)
print(pi_est)
```
